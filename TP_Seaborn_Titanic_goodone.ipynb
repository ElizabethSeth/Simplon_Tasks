```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

* survival: un  Boolean indiquant si le passager a survécu ou non (0 = No; 1 = Yes); => target
* pclass: Passenger class (1 = 1st; 2 = 2nd; 3 = 3rd)
* sex: male/female
* age: Age
* sibsp: Nombre de frères et sœurs/conjoints
* parch: Nombre de parents/enfants à bord
* fare: Tarif passager (livre sterling)..
* embarked: Port d’embarquement (C = Cherbourg; Q = Queenstown; S = Southampton)
* class: la classe de voyage


```python
#It indicates that the plot will be a line plot that shows the relationship between age and fare for each group.
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

sns.relplot(x="age", y="fare", col="pclass",
            hue="sex", style="sex",
            kind="line", data=titanic)
plt.show()
```
