```python
with open('connexion.log', 'r') as file:
    lines = file.readlines()
    log_list = []
    utilisateurs = set()
    for line in lines:
        parts = line.split(";")
        if len(parts) >= 2:
            log_list.append(parts)
            login = parts[1]
            utilisateurs.add(login)

with open('utilisateurs!.txt', 'w') as f:
    for utilisateur in sorted(utilisateurs):
        f.write(f"{utilisateur}\n")
        #print(f"{utilisateur}\n")
        
```

```python
# Trouver les utilisateurs qui se sont connectés en dehors des heures de bureau (avant 8h ou après 19h) et enregistrer leurs informations.
suspect = []
for line in log_list: 
    if len(line) >= 3:
        time = line[2].strip().split()[1]
        time_in_int = int(time.split(":")[0])
        if time_in_int < 8 or time_in_int > 19:
            suspect.append([line[1], line[0]])
print(f"lenght of list with suspect ppl is {len(suspect)}\n")


dangerous_ip = []
with open("warning.txt", 'r') as f:
    for line in f:
        dangerous_ip.append(line.strip())
print(dangerous_ip)
```

```python
# Trouver les utilisateurs qui se sont connectés aux IP dangereuses et enregistrer ces informations.
dangerous_ip_set = set(dangerous_ip)
list_of_attackers = []
for record in log_list:
    if record[0] in dangerous_ip_set:
        list_of_attackers.append([record[0], record[1]])
#print(len(list_of_attackers))

#Objectif : Compter les occurrences de connexions aux IP dangereuses par utilisateur et enregistrer ces informations dans un fichier.
attack_counts = {}
for attacker in list_of_attackers:
    user = attacker[1]
    if user in attack_counts:
        attack_counts[user] += 1
    else:
        attack_counts[user] = 1

sorted_attack_counts = sorted(attack_counts.items(), key=lambda kv: kv[1], reverse=True)

with open('suspect!.txt', 'w') as f:
    f.write(f"{'Utilisateur':<19} {'Nombre de connexions à des IP dangereuses':<10}\n")
    f.write(f"{'-'*15} {'-'*45}\n")
    for user, count in sorted_attack_counts:
        f.write(f"{user:<20} {count:<10}\n")
        print(f"{user:<20} {count:<10}\n")
```
